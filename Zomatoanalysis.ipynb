```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe=pd.read_csv("Zomato data .csv")
```

```python
dataframe
type("rate")
```

converting the data type of column rate

```python
def marks(x):
    x=str(x).split('/')
    x=x[0]
    return float(x)
    
   
dataframe['rate']=dataframe['rate'].apply(marks)
print(dataframe.head())
```

```python
ans=dataframe["listed_in"]
sns.countplot(ans)
plt.xlabel("count")
```

```python
x=dataframe.groupby("listed_in")["votes"].sum()
result= pd.DataFrame(x)

plt.plot(result,c="green",marker="o")
plt.xlabel("types of restaurant")
```
